```python
URL = "http://www.history.navy.mil/research/histories/ship-histories/danfs/n/naugatuck-i.html"
import requests
from bs4 import BeautifulSoup
from lxml import html
#for index, row in df.iterrows():
#page = requests.get(row.iloc[0])
page = requests.get(URL)
```

```python
zz = dict()
for index, row in df.iterrows():
    page = requests.get(row.iloc[1])
    tree=html.fromstring(page.content)
    #elements = tree.xpath("//div[@class='body parsys']/div[@class='text parbase section']/p[contains(text(),': t.')]/text()")
    #elements = tree.xpath("//div[@class='bodyContainer']//div[@class='body parsys']/div[@class='text parbase section']/p[contains(text(),': t.')]/text()")
    elements = tree.xpath("//div[@class='text parbase section' and contains(text(),': t.')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),': t.')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'beam')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),': T.')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),': dp.')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),': T')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),': T.')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),': tonnage ')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),': tonnage ')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),': t ')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'StSlp: t')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),': dp.')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),':- t.')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'tons;')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),'tons;')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),': displacement')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),'Gbt:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'Gbt:')]/text()")   
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'Mon:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),'Mon:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'Str:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),'Str:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'Tug:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),'Tug:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),'oat:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'oat:')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section' and contains(text(),'Slp')]/text()")
    if (elements==[]):
        elements =tree.xpath("//div[@class='text parbase section']/p[contains(text(),'Slp')]/text()")
    zz[index]=elements
    print(f'{index}: {elements}')
#soup = BeautifulSoup(page.content, "html.parser")
#job_elements = soup.find_all("div", attrs={'class':'text parbase section'})[0]
#print(f'exception: {index}')
```

```python
zzz = pd.DataFrame()
temp= pd.DataFrame()
for index, row in df.iterrows():
    nm = index,df.iloc[index,0]
    st = zz[index]
    try:
        nm = nm[1]
        st = st[0]
    except:
        st=''
    q  = pd.DataFrame({"Name":[nm],"Data":[st]})
    temp=temp.append(q)
temp.reset_index(drop=True)
```
